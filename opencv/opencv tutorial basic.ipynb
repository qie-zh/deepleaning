## 导入库

```python
import cv2 as cv #颜色通道是bgr
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv.imread('Photos/cat.jpg')

def show_img(img):
     plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
     plt.show()
show_img(img)
```

```python
#高，宽，颜色通道
img.shape
```

```python
img_gray = cv.imread('Photos/cat.jpg',cv.IMREAD_GRAYSCALE)
show_img(img_gray)
```
